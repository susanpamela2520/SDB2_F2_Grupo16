```python
import pandas as pd

file_path = r"C:\Users\PC\Desktop\bases2p1\title.akas.tsv"

# Leer el archivo TSV
df = pd.read_csv(file_path, delimiter='\t', nrows=1000)

print("Información del DataFrame:")
print(f"Número de filas: {df.shape[0]}")
print(f"Número de columnas: {df.shape[1]}")
print(f"\nNombres de las columnas:")
print(df.columns.tolist())

# Configurar pandas para mostrar todas las filas y columnas
pd.set_option('display.max_rows', None)  # Mostrar todas las filas
pd.set_option('display.max_columns', None)  # Mostrar todas las columnas
pd.set_option('display.width', None)  # Sin límite de ancho
pd.set_option('display.max_colwidth', None)  # Sin límite de ancho de columna

# Ahora mostrar todas las 1000 filas
print("\nTodas las 1000 filas:")
print(df)
```
